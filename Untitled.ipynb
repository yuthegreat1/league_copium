```python
league_data = pd.read_csv("Resources/output.csv")
league_data.head()
```

```python
league_keys = league_data.keys()
league_keys[:30]
```

```python
names = ["Aitori","RyGuyL","Dracotron", "IrenehCp", "BitFish" ]
aitori_list = []
ryguyl_list = []
dracotron_list = []
irenehcp_list = []
bitfish_list = []
game_list = []
for key in league_keys:
    nameless_key = True
    for name in names:
        if name in key and name == "Aitori":
            aitori_list.append(key)
            nameless_key = False
        if name in key and name == "RyGuyL":
            ryguyl_list.append(key)
            nameless_key = False
        if name in key and name == "Dracotron":
            dracotron_list.append(key)
            nameless_key = False
        if name in key and name == "IrenehCp":
            irenehcp_list.append(key)
            nameless_key = False
        if name in key and name == "BitFish":
            bitfish_list.append(key)
            nameless_key = False
    if nameless_key == True:
        game_list.append(key)
            #for all the names if non of them are in the key, add it to game list
```
